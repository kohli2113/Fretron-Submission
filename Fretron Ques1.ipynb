```python
import matplotlib.pyplot as plt

def routes(flights):
    plt.figure()
    for i, flight in enumerate(flights):
        x, y = zip(*flight)
        plt.plot(x, y, label=f'Flight {i+1}')
    plt.xlabel('X coordinate')
    plt.ylabel('Y coordinate')
    plt.title('Flight Paths')
    plt.legend()
    plt.grid(True)
    plt.show()

flights = [[(1, 1), (2, 2), (3, 3)],  #flight 1
           [(1, 1), (2, 4), (3, 2)],  #flight 2
           [(1, 1), (4, 2), (3, 4)]   #flight 3
          ]

routes(flights)
```
